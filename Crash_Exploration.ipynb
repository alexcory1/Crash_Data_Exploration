```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Crash_Vehicle_Data_(SOR).csv')
```

```python
df
```

```python
df = df.dropna(subset=['MAKE', 'VYEAR', 'CRASH_YEAR', 'FATALITIES', 'DRIVERAGE'])
df = df[df['MAKE'].str.upper() != 'NR']
df = df[df['VYEAR'] != 7777]
df['MAKE'] = df['MAKE'].str.title()
```

```python
make_counts = df['MAKE'].value_counts()
top_50_makes = make_counts.head(50)
make_counts_df = top_50_makes.reset_index()
make_counts_df.columns = ['MAKE', 'COUNT']
make_counts_df['LOG_COUNT'] = np.log(make_counts_df['COUNT'])
```

```python
car_type = sns.barplot(data = make_counts_df, x = 'MAKE', y = 'COUNT')
plt.xticks(rotation=-45, fontsize = 8)
plt.xlabel('Make')
plt.ylabel('Count')
plt.title('Number of Crashes by Car Manufacturer')
plt.show()
```

```python
df['VEHICLE_AGE'] = df['CRASH_YEAR'] - df['VYEAR'] + 1 #Cars are released a year in advance, ie 2025 cars come out in 2024
df = df[df['VEHICLE_AGE'] < 35]
df = df[df['VEHICLE_AGE'] > 0] #Anomaly showing -182 years old
age_counts = df['VEHICLE_AGE'].value_counts().sort_index()
age_counts_df = age_counts.reset_index()
age_counts_df.columns = ['VEHICLE_AGE', 'COUNT']
sns.barplot(x='VEHICLE_AGE', y='COUNT', data=age_counts_df)
plt.xlabel('Vehicle Age')
plt.ylabel('Number of Crashes')
plt.show()
```

```python
df_top_50 = df[df['MAKE'].isin(top_50_makes)]
```

```python
top_50_makes = make_counts.head(50).index
df_top_50 = df[df['MAKE'].isin(top_50_makes)]
fatalities_by_make = df_top_50.groupby('MAKE')['FATALITIES'].sum().reset_index()
fatalities_by_make = fatalities_by_make.sort_values(by='FATALITIES', ascending=False)
sns.barplot(x='FATALITIES', y='MAKE', data=fatalities_by_make)
plt.xlabel('Number of Fatalities')
plt.ylabel('Make')
plt.title('Number of Fatalities by Make (Top 50 Makes)')
plt.tight_layout()
plt.show()
```
